```python
landmean = df['Landsize'].mean()
print(landmean)
```

```python
df['Landsize'] = df['Landsize'].fillna(landmean)
```

```python
areamean = df['BuildingArea'].mean()
print(areamean)
```

```python
df['BuildingArea'] = df['BuildingArea'].fillna(areamean)
```

```python
df['Car'] = df['Car'].fillna(carmean)
```

```python
propertymean =  df['Propertycount'].mean()
print(propertymean)
```

```python
print(df['CouncilArea'].value_counts())
```

```python
df['CouncilArea'] = df['CouncilArea'].fillna(0)
df['Regionname'] = df['Regionname'].fillna(0)
```

```python
print(df.isna().sum())
```

```python
newdf = df.drop(columns = ['Longtitude','Lattitude','YearBuilt','Address','Postcode','Date'])
newdf.head()
```

```python
col = ['Suburb','Rooms','Type','Method','SellerG','Distance','Bedroom2','Bathroom','Car','Landsize','BuildingArea','CouncilArea','Regionname','Propertycount','Price']
newdf1 = newdf[col]
newdf1.head()
```

```python
newdf2 = pd.get_dummies(newdf1 ,dtype = int)
```

```python
x = newdf2.drop(columns = ['Price'])
y= newdf2['Price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
print(newdf2.isna().sum())
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.linear_model import Lasso
model1 = Lasso(alpha = 25,max_iter=100,tol=0.1)
model1.fit(x_train,y_train)
```

```python
model1.score(x_test,y_test)
```
